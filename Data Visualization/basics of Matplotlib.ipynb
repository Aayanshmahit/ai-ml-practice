```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
plt.plot(years,runs)
```

```python
plt.plot(years,runs)
plt.xlabel("Runs scored")
plt.ylabel("Years")
plt.title("Sachin Materblaster")
```

```python
plt.style.use('seaborn-v0_8-bright')
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
years = [2010,2011,2012,2013,2014,2015,2016,2017,2018,2019]
plt.plot(years,kohli,label = "kohli",color="yellow",linestyle='--',linewidth=3)
plt.plot(years,sehwag,label = "sehwag",color="red",linestyle='-',linewidth=2)
plt.xlabel("Years")
plt.ylabel("Runs sored")
plt.title("comparing kohli and sehwag")
plt.legend()
plt.grid(True)
```

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
years = [2010,2011,2012,2013,2014,2015,2016,2017,2018,2019]
with plt.xkcd():
    plt.plot(years,kohli,label = "kohli")
plt.plot(years,sehwag,label = "sehwag")
plt.xlabel("Years")
plt.ylabel("Runs sored")
plt.title("comparing kohli and sehwag")
plt.legend()
```
