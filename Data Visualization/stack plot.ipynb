```python
import matplotlib.pyplot as plt
```

```python
activities = ['Studying', 'Playing', 'Watching TV', 'Sleeping']
time_spent = [3, 2, 2, 5]
colours = ['Red', 'Blue','Green','Yellow']
plt.figure(figsize = (6,6))
plt.pie(time_spent , labels = activities,autopct = '%1.1f%%',startangle = 90)
```

```python
days = [1, 2, 3, 4, 5, 6, 7]
studying = [3, 4, 3, 5, 4, 3, 4]
playing = [2, 2, 1, 1, 2, 3, 2]
watching_tv = [2, 1, 2, 2, 1, 1, 1]
sleeping = [5, 5, 6, 5, 6, 5, 5]
labels = ['Studying', 'Playing', 'Watching TV', 'Sleeping']
plt.stackplot(days,studying , playing , watching_tv,sleeping , labels = labels,alpha = 0.7)
plt.legend(loc = "upper right")
plt.grid()
plt.xlabel("days")
plt.ylabel("Activities")
plt.title("tracking the Activities")
```
