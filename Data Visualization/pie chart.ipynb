```python
import matplotlib.pyplot as plt
```

```python
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
colors = ['#9C2005','#66b3ff','#99ff99','#ffcc99']
explode = [0.2 , 0 , 0 , 0]
plt.pie(runs ,labels = labels,colors = colors
        ,wedgeprops = {'edgecolor':"black",'linewidth' : 0.5 , 'linestyle' : "-"},shadow = True,
       explode = explode, autopct = '%1.1f%%')
plt.title("Greatest of all time")
```
